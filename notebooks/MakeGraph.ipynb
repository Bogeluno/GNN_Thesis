```python
def haversine_alt(df, car1, car2):
    dfc1 = df[df.car == car1]
    dfc2 = df[df.car == car2]

    point1 = dfc1[['park_location_lat','park_location_long']].values[0]
    point2 = dfc2[['park_location_lat','park_location_long']].values[0]
    #return point1

    # convert decimal degrees to radians
    lon1, lat1 = map(np.radians, point1)
    lon2, lat2 = map(np.radians, point2)

    # Deltas
    delta_lon = lon2 - lon1 
    delta_lat = lat2 - lat1 
    
    # haversine formula 
    a = np.sin(delta_lat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(delta_lon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371000 # Radius of earth in m
    return c * r
```
